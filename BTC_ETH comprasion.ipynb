```python
import quandl
import pandas as pd
btceur=quandl.get("BITFINEX/BTCEUR", collapse="monthly", start_date="2018-03-18")
ethusd=quandl.get("BITFINEX/ETHUSD", collapse="monthly", start_date="2018-03-18")
```

```python
import matplotlib.pyplot as plt

plt.style.use("dark_background")
plt.figure(figsize=(20,7))
plt.grid(linewidth=0.2)
plt.title("BTC/ETH (USD)")
plt.xlabel('Date')
plt.ylabel('USD')
plt.plot(btceur.index, btceur.Last)
plt.plot(ethusd.index, ethusd.Last)
```
